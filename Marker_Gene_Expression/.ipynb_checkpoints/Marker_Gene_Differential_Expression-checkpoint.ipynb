```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from matplotlib.lines import Line2D
```

```python
metagenome = 'SRX247406'
metatranscriptome = 'SRR769418'
```

Part 1: Append OTU to Marker Gene Table

This uses the linkage group table found in 'mOTUs_v2/db_mOTU' downloaded with mOTUs_v2 to append the OTU to each marker gene. The marker gene tables ('_mgc.txt') are generated by running motus profile with -M option. This is done for a metagenome and metatranscriptome originating from the same sample. 


```python
gene_to_mOTU_table = pd.read_csv('mOTU-LG.map.tsv',sep='\t',names=['OTU'],index_col=0)

genome_output = open(metagenome + '_MGC_OTUs.txt','w')
mgc_table = open(metagenome + '_mgc.txt')
genome_output.write(mgc_table.next())
genome_output.write(mgc_table.next())

for line in mgc_table:
    gene = line.split('\t')[0]
    OTU = gene_to_mOTU_table.loc[gene,'OTU']
    genome_output.write(line.replace('\n','\t' + OTU + '\n'))

genome_output.close()
```

```python
transcriptome_output = open(metatranscriptome + '_MGC_OTUs.txt','w')
mgc_table = open(metatranscriptome + '_mgc.txt')
transcriptome_output.write(mgc_table.next())
transcriptome_output.write(mgc_table.next())

for line in mgc_table:
    gene = line.split('\t')[0]
    OTU = gene_to_mOTU_table.loc[gene,'OTU']
    transcriptome_output.write(line.replace('\n','\t' + OTU + '\n'))

transcriptome_output.close()
```

Part 2: Make table with OTUs as index and column for each marker gene

```python
expression_matrix = {}
for data in [metagenome, metatranscriptome]:
    gene_abundance_mOTU_table = pd.read_csv(data + '_MGC_OTUs.txt',sep='\t',header=1,names=['scaled_counts','OTU'])
    gene_list = []
    for name in gene_abundance_mOTU_table.index:
        gene_name = name.split('.')[0]
        gene_list.append(gene_name)

    gene_abundance_mOTU_table['gene'] = gene_list

    repeat = 0
    new = 0
    OTUs = []
    for OTU in gene_abundance_mOTU_table['OTU']:
        if OTU in OTUs:
            repeat = repeat + 1
        else: 
            OTUs.append(OTU)
            new = new + 1    

    OTU_by_MG = pd.DataFrame(OTUs)
    OTU_by_MG.columns = ['OTU']
    OTU_by_MG.index = OTU_by_MG['OTU']
    OTU_by_MG = OTU_by_MG.drop('OTU',axis=1)
    OTU_by_MG = OTU_by_MG.drop(['-1'])

    for marker_gene in gene_abundance_mOTU_table.groupby('gene').sum().index:
        sub_table = gene_abundance_mOTU_table[gene_abundance_mOTU_table['gene']==marker_gene]
        sub_table.columns = [marker_gene,'OTU','gene']
        sub_table.index = sub_table['OTU']
        OTU_by_MG = OTU_by_MG.join(sub_table[marker_gene], lsuffix = 'l',rsuffix = 'r', how = 'left')

    OTU_by_MG = OTU_by_MG.fillna(value=0)
    #OTU_by_MG.to_csv(data + '_MG_matrix.txt',sep='\t')
    expression_matrix[data] = OTU_by_MG
```

Append species to each to make the matrix qualitatively more informative; uses a table of taxonomic names for each OTU found in 'mOTUs_v2/db_mOTU/'

```python
OTU_list = open('CAMI_Sep_2015_mOTUs_2.0.0.tsv')
OTU_list.next()
OTU_list.next()
ID_to_phylum = {}
for line in OTU_list:
    ID = line.split('\t')[0]
    taxonomy = line.split('\t')[28]
    phylum = taxonomy.split('|')[1]
    Class = taxonomy.split('|')[2]
    order = taxonomy.split('|')[3]
    family = taxonomy.split('|')[4]
    genus = taxonomy.split('|')[5]
    species = taxonomy.split('|')[6]
    ID_to_phylum[ID] = phylum

for data in [metagenome,metatranscriptome]:
    phylum_list = []
    for ID in expression_matrix[data].index:
        if isinstance(ID,basestring):
            phylum_list.append(ID_to_phylum[ID])
        else:
            phylum_list.append('NA')
    tuples = list(zip(expression_matrix[data].index,phylum_list))
    index = pd.MultiIndex.from_tuples(tuples, names=['ID', 'phylum'])
    expression_matrix[data].index = index
```

Part 3: Check correlation of marker genes to each other to see if there is any differential expression of marker genes

```python
expression_matrix[metagenome]+=0.0001
expression_matrix[metatranscriptome]+=0.0001

expression_matrix_log={}
expression_matrix_log[metagenome] = np.log10(expression_matrix[metagenome])
expression_matrix_log[metatranscriptome] = np.log10(expression_matrix[metatranscriptome])
```

```python
f,(ax1, ax2) = plt.subplots(1,2,sharey=True)

expression_matrix_log[metagenome].plot(x='COG0012',y='COG0172',style='o',logx=True,logy=True,ax=ax1,legend=False)
expression_matrix_log[metatranscriptome].plot(x='COG0012',y='COG0172',style='o',logx=True,logy=True,ax=ax2,legend=False)

ax1.set(ylabel='COG0172',title='metagenome')
ax2.set(title='metatranscriptome')
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, gridspec_kw={'width_ratios': [5,5,0.2]},figsize=(10,5))
sns.heatmap(expression_matrix_log[metagenome].corr(method='spearman'),vmin=0.5,vmax=1,ax=ax1,cbar=False)
sns.heatmap(expression_matrix_log[metatranscriptome].corr(method='spearman'),vmin=0.5,vmax=1,ax=ax2,yticklabels=False,cbar_ax=ax3)
ax1.set_title('Metagenome Marker Gene \n Spearman Correlations (log)',size=10)
ax2.set_title('Metatranscriptome Marker Gene \n Spearman Correlations (log)',size=10)
```

Below we tried to plot gene levels for members of a given phylum but this ends up being too sparse to be informative

```python
expression_matrix_log[metagenome] = expression_matrix_log[metagenome].sort_values(by=['COG0012'],ascending=False)
expression_matrix_log[metatranscriptome] = expression_matrix_log[metatranscriptome].sort_values(by=['COG0012'],ascending=False)

Actinobacteria_genome = expression_matrix_log[metagenome].xs('Actinobacteria',level=1)
Actinobacteria_transcriptome = expression_matrix_log[metatranscriptome].xs('Actinobacteria',level=1)

f, (ax1, ax2) = plt.subplots(1, 2,figsize=(20,10))

for gene in Actinobacteria_genome.columns:
    Actinobacteria_genome.plot(y=gene,style='o',ax=ax1, legend=False,rot=90)

for gene in Actinobacteria_transcriptome.columns:
    Actinobacteria_transcriptome.plot(y=gene,style='o',ax=ax2, legend=False,rot=90)

ax1.set_xlabel('')
ax2.set_xlabel('')
```

To fix problem of sparsity, we looked only the most abundant OTUs based on the profile of OTUs generated by mOTUs2 and then plotted the abundances of each gene to look at the variation in the metagenome vs metatranscriptome

```python
metagenomic_profile = pd.read_csv(metatranscriptome + '_profile.txt',header = 2,sep='\t')
metagenomic_profile.columns = ['OTU','Abundance']
metagenomic_profile = metagenomic_profile.sort_values('Abundance',ascending = False)
most_abundant = metagenomic_profile.iloc[0:9]['OTU']
top_IDs = []
for name in most_abundant:
    ID = name.split('[')[1]
    ID = ID.split(']')[0]
    top_IDs.append(ID)
```

```python
idx = pd.IndexSlice
metagenome_top_OTUs = expression_matrix_log[metagenome].loc[idx[top_IDs,:],:]
metatranscriptome_top_OTUs = expression_matrix_log[metatranscriptome].loc[idx[top_IDs,:],:]
```

```python
f, (ax1, ax2) = plt.subplots(1, 2,figsize=(24,12))

for gene in metagenome_top_OTUs.columns:
    metagenome_top_OTUs.plot(y=gene,style='o',ms=10,ax=ax1, legend=False,rot=90)
    metatranscriptome_top_OTUs.plot(y=gene,style='o',ms=10,ax=ax2, legend=False,rot=90)

ax1.set_xlabel('')
ax1.set_ylabel('log10(Relative Expression)')
ax2.set_xlabel('')
ax2.set_ylabel('log10(Relative Expression)')

ax1.set_title('Metagenome Per Marker Gene Expression Levels, Most Abundant OTUs in Metatranscriptome')
ax2.set_title('Metatranscriptome Per Marker Gene Expression Levels, Most Abundant OTUs in Metatranscriptome')
```

Below we used the full matrix of expression levels (including all OTUs) to look at PCA plots of each OTU. We asked if there was any stratification for ref vs meta OTUs, which would indicate a technical bias in the method. We then looked at any stratification by phylum to see if they might have similar expression patterns. We did not see any clear pattern in either case

Reference for PCA code: https://towardsdatascience.com/pca-using-python-scikit-learn-e653f8989e60

```python
expression_matrix_log[metatranscriptome]['target'] = expression_matrix_log[metatranscriptome].index
```

```python
features = list(expression_matrix_log[metatranscriptome].columns.drop(['target']))
x = expression_matrix_log[metatranscriptome].loc[:,features].values
y = expression_matrix_log[metatranscriptome].loc[:,['target']].values

x = StandardScaler().fit_transform(x)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
target_series = expression_matrix_log[metatranscriptome]['target']
target_series.index = range(0,len(target_series))
finalDf = pd.concat([principalDf, expression_matrix_log[metatranscriptome]['target']], axis = 1)
```

```python
ref_meta = []
for target in finalDf['target']:
    if isinstance(target[0],basestring):
        temp = target[0].split('_')[0]
        if temp == 'meta':
            value = 'r'
        if temp == 'ref':
            value = 'b'
        ref_meta.append(value)
    else:
        ref_meta.append('g')
finalDf['ref_meta'] = ref_meta
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('10 Marker Gene PCA', fontsize = 20)

legend_elements = [Line2D([0], [0], marker='o', color='w', label='meta',
                          markerfacecolor='r', markersize=10),
                  Line2D([0], [0], marker='o', color='w', label='ref',
                          markerfacecolor='b', markersize=10),
                  Line2D([0], [0], marker='o', color='w', label='na',
                          markerfacecolor='g', markersize=10)]

ax.legend(handles=legend_elements, loc='right')

ax.scatter(finalDf['principal component 1'], finalDf['principal component 2'], color = finalDf['ref_meta'])
```

```python
phylum_list = []
for item in finalDf['target']:
    phylum_list.append(item[1])
finalDf['phylum'] = phylum_list
```

```python
sns.lmplot(x="principal component 1", y="principal component 2", data=finalDf, fit_reg=False, hue='phylum', legend=True)
ax = plt.gca()
ax.set_title("10 Marker Gene PCA")

```

To use the expression matrix in edgeR, the code below normalizes the metatranscriptome abundances based on the mOTU abundances in the metagenome profile. Then we proceeded to run the R code in this folder. 

```python
metagenome = 'SRX247406'
metatranscriptome = 'SRR769418'

expression_matrix = pd.read_csv(metatranscriptome + '_expression_matrix.txt',header=0,sep='\t',index_col=0)
profile = pd.read_csv(metagenome + '_profile.txt',header=2,sep='\t')
profile.columns = ['OTU_full','genome_abundance']

list_for_index = []
for entry in profile['OTU_full']:
    if entry != '-1':
        new = entry.split('[')[1]
        new = new.split(']')[0]
        list_for_index.append(new)
    else:
        list_for_index.append('-1')

profile.index = list_for_index

merged = expression_matrix.join(profile['genome_abundance'], lsuffix = 'l',rsuffix = 'r', how = 'left')

for column in merged.columns:
    if column != 'phylum':
        expression_matrix[column] = merged[column]/merged['genome_abundance']
    else:
        expression_matrix[column] = merged[column]

expression_matrix = expression_matrix[expression_matrix['genome_abundance']>0]
expression_matrix = expression_matrix.drop('genome_abundance',axis=1)

expression_matrix.to_csv(metatranscriptome + '_normalized_matrix.txt',sep='\t')
```
