```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gene_abundances = open('SRR769418_kneaddata_genefamilies.tsv')
species_count = {}
for line in gene_abundances:
    if 'UNMAPPED' in line:
        species_count['unmapped'] = float(line.split('\t')[1])
    if '|' in line:
        info = line.split('|')[1]
        species = info.split('\t')[0]
        if 's__' in species:
            species = species.split('__')[2]
        count = float(info.split('\t')[1])
        if species in species_count:
            species_count[species] = species_count[species] + count
        else:
            species_count[species] = count
```

```python
plt.bar(range(len(species_count)), list(species_count.values()), align='center', log=True)
plt.xticks(range(len(species_count)),list(species_count),rotation=90)
plt.title('Total Transcript Abundance by Species')
plt.ylabel('Abundance (reads)')
```
