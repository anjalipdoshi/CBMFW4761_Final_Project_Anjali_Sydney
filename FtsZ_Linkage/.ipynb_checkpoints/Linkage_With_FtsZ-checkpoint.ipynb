To try to add ftsZ to mOTUs linkage groups, we used ftsZ orthologs as a reference for an mOTUs profile (within the MOCAT implementation). To do this, we had to replace the COG0099 file with ftsZ ortholog sequences. The output of the profiling is a bam file, which we converted to a sam file and used below to extract counts of each ortholog.

```python
sam = open('filtered.COG0099.sam')
output = open('simulated_87_COG0099_counts.txt','w')
gene_to_count = {}
for line in sam:
    gene = line.split('\t')[2]
    if gene in gene_to_count:
        gene_to_count[gene] += 1
    else:
        gene_to_count[gene] = 1
for gene in gene_to_count:
    output.write(gene + '\t' + str(gene_to_count[gene]) + '\n')
output.close()
```

We then used the counts of ftsZ to try to make linkage groups by looking at the correlation with the OTU abundances generated by a standard implementation of mOTUs. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
profile = pd.read_csv('gutSimulation_profile.txt',sep='\t',header=2)
profile.columns = ['OTU_full','abundance']
```

```python
list_for_index = []
for entry in profile['OTU_full']:
    if entry != '-1':
        new = entry.split('[')[1]
        new = new.split(']')[0]
        list_for_index.append(new)
    else:
        list_for_index.append('-1')

profile.index = list_for_index
```

```python
profile_non_zero = profile[profile['abundance']>0]
```

```python
profile_values = profile_non_zero.sort_values('abundance',ascending=False)['abundance']
```

```python
ftsZ = pd.read_csv('simulated_87_COG0099_counts.txt',sep='\t',header=None)
ftsZ.columns = ['name','abundance']
```

```python
ftsZ_values = ftsZ.sort_values('abundance',ascending=False)['abundance']
```

```python
ftsZ_values.index = range(0,6)
ftsZ_values
```

```python
ftsZ_add_zeros = []
for n in range(0,6):
    ftsZ_add_zeros.append(ftsZ_values[n])
for n in range(6,87):
    ftsZ_add_zeros.append(0)
```

```python
profile_list = []
for n in range(0,87):
    profile_list.append(profile_values[n])
```

```python
data = pd.DataFrame(ftsZ_add_zeros,profile_list)
data = pd.DataFrame(list(zip(ftsZ_add_zeros, profile_list)), 
               columns =['ftsZ', 'OTUs']) 
```

```python
data['ftsZ'] = data['ftsZ']/np.sum(data['ftsZ'])
data['OTUs'] = data['OTUs']/np.sum(data['OTUs'])
```

```python
plt.scatter(data['OTUs'],data['ftsZ'])
plt.xlabel('genome abundance')
plt.ylabel('ftsZ abundance')
plt.title('Attempted Linkage Group Formation for ftsZ')
```
